# Για τα αεροπορικά ατυχήματα στη χώρα μας σύμφωνα με τις εκθέσεις της Επιτροπής Διερεύνησης τους


*Έρευνα-Ρεπορτάζ : Ειρήνη Δερμιτζάκη – Σωτήρης Τάρλας*

Το ανθρώπινο λάθος και εν τέλει κυρίως αυτό του χειριστή κρύβεται πίσω από τα σοβαρά (και όχι μόνον) αεροπορικά ατυχήματα της Πολιτικής Αεροπορίας όπως προκύπτει από τις εκθέσεις /πορίσματα της καθ’ ύλην αρμόδιας Επιτροπής Διερεύνησης τους.
Η ΕΔΑΑ (Επιτροπή Διερεύνησης Αεροπορικών Ατυχημάτων) έχει ως έργο τη διερεύνηση ασφαλείας όλων των αεροπορικών ατυχημάτων και σοβαρών συμβάντων της πολιτικής αεροπορίας που σημειώνονται τόσο εντός της Περιοχής Πληροφοριών Πτήσεων (FIR) Αθηνών όσο και εκτός όταν πρόκειται για αεροσκάφος Ελληνικού Νηολογίου η το εκμεταλλεύεται επιχείρηση με έδρα την Ελλάδα εφόσον η διερεύνηση τους δεν διενεργείται από άλλο κράτος. Η Επιτροπή που συστήθηκε το 2001 όταν η Ελληνική νομοθεσία προσαρμόστηκε με τον νόμο 2912 στην από το 1994 Οδηγία 94/56 της Ευρωπαϊκής Κοινότητας (την οποία έχει καταργήσει ο Νέος Κανονισμός της Ευρωπαϊκής Ένωσης 996/2010 που ισχύει μέχρι σήμερα) έχει διερευνήσει τα σχεδόν 200 ατυχήματα ή σοβαρά συμβάντα από την αυγή της Νέας Χιλιετίας. Τα οποία καταμερίζονται στις εκθέσεις/πορίσματα της σε έξι βασικές κατηγορίες. Συμβάν, σοβαρό συμβάν , ατύχημα χωρίς, με μικρό, σοβαρό ή θανάσιμο τραυματισμό.
Ανάμεσα στα 197 συμβάντα - ατυχήματα περιλαμβάνονται και 24 θανατηφόρα με έναν ή περισσότερους νεκρούς.


### Στο Γραμματικό το 64% των θυμάτων  

Συνολικά 189 άτομα (64 μέλη πληρώματος και 125 επιβάτες) έχασαν τη ζωή τους σ’ αυτό το διάστημα των σχεδόν 22 ετών από τη σύσταση της Επιτροπής με την πλειοψηφία των θυμάτων να βρίσκονται στη λίστα εκείνων της τραγωδίας της 14ης Αυγούστου 2005 στο Γραμματικό όταν το Boeing της αεροπορικής εταιρείας «Helios Airways» συνετρίβη, σε λοφώδη περιοχή πλησίον της κωμόπολης  της Βορειανατολικής Αττικής κοντά στο Μαραθώνα ενώ εκτελούσε πτήση από τη Λάρνακα στην Πράγα με ενδιάμεσο σταθμό το αεροδρόμιο «Ελευθέριος Βενιζέλος», οδηγώντας στο θάνατο 121 ανθρώπους. Τα έξι μέλη του πληρώματος και τους 115 επιβάτες. Όπως αναφέρει το τελικό πόρισμα  της 4ης Οκτωβρίου 2006 αιτία του μοιραίου συμβάντος ήταν η αποσυμπίεση της καμπίνας που είχε ως αποτέλεσμα την αδυναμία χειρισμού του αεροπλάνου από το πλήρωμα διακυβέρνησης λόγω υποξίας (στέρησης επαρκούς οξυγόνωσης του ανθρώπινου σώματος). Η αποσυμπίεση στο εσωτερικό της καμπίνας  που εκδηλώθηκε λίγη ώρα μετά την απογείωση ήταν ταχεία, δεν υπήρχε καθόλου οξυγόνο μ’ αποτέλεσμα να λιποθυμήσουν όλοι οι επιβαίνοντες εκτός από τον αεροσυνοδό Ανδρέα Προδρόμου ο οποίος είχε πτυχίο πιλότου για μικρότερα αεροπλάνα. Πήρε τον έλεγχο του μοιραίου αεροσκάφους που όμως παρά τις προσπάθειες του έκτακτου χειριστή κι αφού έκανε αρκετούς κύκλους πάνω από την Αττική , έμεινε τελικά από καύσιμα για να ακολουθήσει η συντριβή στο Γραμματικό λίγα λεπτά μετά τις 12 το μεσημέρι.


### Οι δύο τραγωδίες του ΕΚΑΒ
Με εξαίρεση τρία παλαιότερα από εκείνη την τραγωδία, που σόκαρε όχι μόνον το πανελλήνιο και τον κυπριακό ελληνισμό αλλά και είχε ανάλογη διεθνή απήχηση, μόλις τρία από τα υπόλοιπα ατυχήματα στην Ελλάδα είχαν τέσσερα ή περισσότερα θύματα. Ανάμεσα τους και δύο με κοινό στοιχείο τον ΕΚΑΒ. Πρώτα αυτό της 14ης Ιανουαρίου 2001 όταν στη θαλάσσια περιοχή επτά ναυτικά μίλια νοτίως του Σουνίου ελικόπτερο διακομιδής ασθενούς συνετρίβη σε συνθήκες νύχτας κι έντονων καιρικών φαινομένων και κόστισε τη ζωή και στους πέντε επιβαίνοντες (το 4μελές πλήρωμα και τον μοναδικό επιβάτη).Τον ίδιο ακριβώς απολογισμό είχε και το ατύχημα σχεδόν 1,5 χρόνο μετά (16/6/2002) με άλλο ελικόπτερο του Εθνικού Κέντρου Άμεσης Βοήθειας το οποίο σε νυχτερινή επίσης πτήση κατά την αεροδιακομιδή ασθενούς συνετρίβη σε όρος της νήσου Ανάφης και κατεστράφη ολοσχερώς.
Πέραν των προαναφερθέντων θανατηφόρων ατυχημάτων στο διάστημα της τελευταίας 20ετίας έχουν σημειωθεί και μερικές δεκάδες άλλα μ’ έναν ή περισσότερους σοβαρούς τραυματισμούς.


Η συντριπτική πλειοψηφία των συμβάντων και ατυχημάτων έχει λάβει χώρα είτε εντός είτε πολύ κοντά σε αερολιμένα και τα περισσότερα εξ αυτών με σοβαρό  ή θανάσιμο τραυματισμό  σημειώθηκαν την 5ετία 2010-2015(11). Αντίθετα η λιγότερο αιματοβαμμένη περίοδος είναι από τις 18/7/2015 ως τις 5/10/2016 καθώς δεν υπήρξε ούτε ένα και τη διετία 2005-06 σημειώθηκαν μόλις δύο. Τα περισσότερα (τέσσερα) θανατηφόρα την ίδια χρονιά έχουν συμβεί δύο φορές (2009 και 2014).

Αποκλειστικός σκοπός της διερεύνησης της ΕΔΑΑΠ είναι η εξακρίβωση των αιτίων που προκάλεσαν το ατύχημα ή το σοβαρό συμβάν, με στόχο την πρόληψη ατυχημάτων ή συμβάντων και δεν αποσκοπεί στη διαπίστωση υπαιτιότητας και τον καταλογισμό ευθυνών ή αξιώσεων. Σε διερεύνηση υπόκεινται όλα τα ατυχήματα ή σοβαρά συμβάντα που συνέβησαν σε αεροσκάφος κατά τη διάρκεια της πτητικής λειτουργίας του ή πριν και μετά αυτής εάν η Επιτροπή αναμένει από την έρευνα να προκύψουν νέα δεδομένα για την βελτίωση της ασφάλειας πτήσεων.

### Η ανθρώπινη ενοχή
Μέσα από τις εκθέσεις – πορίσματα της Επιτροπής προκύπτει μια πανταχού παρούσα… ενοχή. Αυτή του ανθρώπινου παράγοντα όπως επισημαίνει – επιβεβαιώνει ο ιπτάμενος εκπαιδευτής Ηλίας Τάρλας. Ο πολύπειρος πιλότος της 359 ΜΑΕΔΥ(Μονάδα Αεροπορικών Εφαρμογών Δημοσίων Υπηρεσιών) η οποία ανήκει στην Πολεμική Αεροπορία, εκπαιδευτής – εξεταστής της Πολιτικής Αεροπορίας με περισσότερες από 8.500 ώρες πτήσης με πυροσβεστικά και όχι μόνον αεροπλάνα και συνεργάτης-διερευνητής της ΕΔΑΑ υπογραμμίζει σχετικά : «Σ’ όλη την αλυσίδα της πτήσης όταν συμβαίνει ένα αεροπορικό ατύχημα έχει περίτρανα αποδειχθεί ότι είναι πάντοτε παρόν το ανθρώπινο σφάλμα. Και αυτό γιατί και στο κομμάτι κατασκευή του αεροσκάφους και στη διεξαγωγή της πτήσης καθώς και σ’ ότι αφορά αυτούς που την υποστηρίζουν (σχέδιο, εναέρια κυκλοφορίας κ.λ.π.) εμπλέκονται άνθρωποι. Συνεπώς το ανθρώπινο σφάλμα είναι αυτό το οποίο καθορίζει το πότε και πως ένα αεροπορικό ατύχημα θα εξελιχτεί». 
Η διερευνητική εμπειρία του εμπεριέχει πέραν των άλλων και τρία θανατηφόρα ατυχήματα ανάμεσα στα οποία και εκείνο με πρωταγωνιστή – θύμα ένα φίλο- μαθητή του όταν στις 29 Απριλίου 2011 σε ιδιωτικό αεροδρόμιο στο Πόρτο Χέλι ελαφρύ αεροπλάνο που εκτελούσε προσγειώσεις – απογειώσεις έπεσε και βυθίστηκε στη θάλασσα σε βάθος εννέα μέτρων και βρήκαν το θάνατο αμφότεροι οι επιβαίνοντες. Για το οποίο ζήτησε εξαίρεση από τη συμμετοχή στην έρευνα για την αποφυγή σφάλματος για λόγους συναισθηματικής φόρτισης την οποία επέτεινε το γεγονός ότι είχε επικοινωνία με τον πιλότο φίλο του λίγα λεπτά πριν από τη μοιραία πτώση του αεροσκάφους όντας και ο ίδιος σε πτήση στην ίδια περιοχή του ατυχήματος.














### Το τελικό σφάλμα o χειριστής
«Ως εργαζόμενος στην Πολεμική Αεροπορία ανήκω στο Γραφείο Ασφάλειας Πτήσεων της Μοίρας μου κι έχω επίσης πρόσβαση σ’ όλες τις διερευνήσεις για τα αεροπορικά ατυχήματα που έχουν συμβεί στη Μοίρα και όχι μόνον (πυροσβεστικά, μεταφορικά, μαχητικά) και είναι διαπιστωμένο ότι η ίσως μοναδική αιτία τους είναι ο ανθρώπινος παράγοντας. Αυτός μπορεί να έχει εμπλοκή είτε στην τεχνική υποστήριξη, στο χειριστικό κομμάτι (εσφαλμένη απόφαση), ιατρικό έλεγχο, έλεγχο εναέριας κυκλοφορίας, ακόμα και σε τρομοκρατική ενέργεια» προσθέτει ο τα τελευταία 12 χρόνια διερευνητής που υπογραμμίζει ότι «δεν αρκεί ένα σφάλμα για να γίνει το ατύχημα, αλλά απαιτούνται τρία και πλέον ορισμένες φορές για να σπάσει η αλυσίδα». Κι εξηγεί : «Ενώ περίπου, όπως προκύπτει μέσα από διεθνείς έρευνες-μελέτες, το 90% όλων των αεροπορικών ατυχημάτων της γενικής και πολιτικής αεροπορίας οφείλονται σε σφάλμα χειριστού και τα περισσότερα δε από τα μισά θανατηφόρα ατυχήματα αποδίδονται σ’ εσφαλμένη απόφαση του δεν είναι από μόνο του αρκετό για να συμβεί το ατύχημα. Και τούτο γιατί το δικό του κρίσιμο λάθος αποδεδειγμένα στηρίζεται σε προηγούμενες εσφαλμένες πληροφορίες ή αστάθμητο παράγοντα. Πρόκειται για αυτό που στην αεροπορική ορολογία ονομάζεται ελβετικό τυρί, όρος που περιγράφει τη διαδοχικότητα των λαθών που συντελούν στην πρόκληση του συμβάντος -ατυχήματος».

### Η υποχρεωτικότητα της δήλωσης ατυχήματος ή συμβάντος
Η κατάθεση δήλωσης με πληροφορίες και στοιχεία ενός αεροπορικού ατυχήματος ή συμβάντος στην ΕΔΑΑΠ είναι υποχρεωτική και με βάση το νόμο γίνεται από τον υπεύθυνο κυβερνήτη ή, αν αυτός κωλύεται, από οποιοδήποτε μέλος του πληρώματος του αεροσκάφους ή, εφόσον κανείς από αυτούς δεν είναι σε θέση, από τον κύριο φορέα της εκμετάλλευσης του αεροσκάφους. Η δήλωση μπορεί γενικά να γίνει και από οποιονδήποτε αντιληφθεί ή λάβει γνώση ατυχήματος, σοβαρού ή μη συμβάντος. Την ίδια υποχρέωση έχουν και οι αρμόδιες για τον έλεγχο, εποπτεία και διευκόλυνση των πτήσεων Αρχές εφόσον λάβουν γνώση ατυχήματος ή σοβαρού συμβάντος που συνέβη κατά τη λειτουργία ενός αεροσκάφους. Η δήλωση συνήθως γίνεται απ’ ευθείας στην ΕΔΑΑΠ αλλά μπορεί επίσης να κατατεθεί στην πλησιέστερη Αεροπορική, Αερολιμενική, Αστυνομική ή άλλη τοπική Αρχή, η οποία πρέπει άμεσα να την ειδοποιήσει. Η κατάθεση της αναφοράς μπορεί να γίνει είτε τηλεφωνικά είτε μέσω διαδικτυακής (στον ιστότοπο της Επιτροπής (https://aaiasb.gr/anafora) ή έντυπης φόρμας. 





### Τα μετά τη δήλωση «βήματα» της ΕΔΑΑΠ


Με την άμεση ειδοποίηση της ΕΔΑΑΠ ο Πρόεδρός της ορίζει Ομάδα Διερεύνησης από τους μόνιμους διερευνητές της Μονάδας και, αν η πολυπλοκότητα του ατυχήματος ή του συμβάντος το επιβάλλει, και συμβούλους με πείρα σε εξειδικευμένους τομείς της πολιτικής αεροπορίας. Οι ειδικότητες των μελών μιας διερευνητικής ομάδας είναι χειριστής, τεχνικός, τηλεπικοινωνίες, γιατρός, νομικός με την προϋπόθεση ότι άπαντες είναι πτυχιούχοι του σχολείου διερεύνησης ατυχημάτων της Πολεμικής Αεροπορίας. Η Ομάδα μεταβαίνει στον τόπο του γεγονότος το ταχύτερο δυνατό με τους διερευνητές σύμφωνα με τον ν.2912/01 να έχουν ελεύθερη πρόσβαση στα στοιχεία (συντρίμματα, απομαγνητοφωνήσεις, εκπαιδευτικούς και ιατρικούς φακέλους εμπλεκομένων, εγχειρίδια, κλπ) τα οποία αφορούν στο υπό διερεύνηση γεγονός, καθώς και δικαίωμα λήψης καταθέσεων από οποιονδήποτε έχει σχέση ή γνωρίζει πληροφορίες σχετικά με αυτό. Οι εργαστηριακές εξετάσεις μηχανών, οργάνων, συγκροτημάτων και εξαρτημάτων των α/φων διεξάγονται σε εξειδικευμένα και κατάλληλα εξοπλισμένα κέντρα του εσωτερικού ή εξωτερικού. Η Ομάδα συλλέγει και εξετάζει όλα τα σχετικά με το ατύχημα ή συμβάν στοιχεία προκειμένου να διαμορφώσει άποψη για τους αιτιογόνους παράγοντες που ενεπλάκησαν στο υπό διερεύνηση γεγονός και η ολοκλήρωση της διαδικασίας οδηγεί στην διατύπωση Συστάσεων Ασφαλείας Πτήσεων.



 «Η Επιτροπή δεν έχει δικούς της πόρους. Χρειάζεται να απλώνει τακτικά κοίλο χέρι προς στο κράτος για οικονομική ενίσχυση . Για τον λόγο αυτό οι διερευνητές είναι εξωτερικοί μη αμειβόμενοι συνεργάτες. Την κάθε επιτροπή τη στελεχώνουν διερευνητές διαφόρων ειδικοτήτων. Ο πρόεδρος της, που είναι ένα πρόσωπο με κύρος και γνώσεις που τυγχάνει ευρείας εκτίμησης από την παγκόσμια αεροπορική κοινότητα , όπως για παράδειγμα ο πρώτος πρόεδρος και νυν αντιπρόεδρος της, αντισμήναρχος ε.α, τέως Κυβερνήτης Πολιτικής Αεροπορίας, Ακριβός Τσολάκης. Επιλέγει τους εκάστοτε συνεργάτες που θα ασχοληθούν με τη διερεύνηση του ατυχήματος. Τα μέλη των επιτροπών αλλάζουν κάθε φορά με εξαίρεση τον πρόεδρο και τον έχοντα το γενικό πρόσταγμα διερευνητή» αναφέρει ο πολύπειρος αεροπόρος και υπογραμμίζει για τα προαπαιτούμενα και τις δυσκολίες της αποστολής των ομάδων διερεύνησης : «Συνήθως ένας από τους διερευνητές πρέπει να είναι χειριστής στον τύπο του αεροπλάνου που εμπλέκεται στο ατύχημα. Το έργο του διερευνητή είναι πολύ συναισθηματικά φορτισμένο γιατί εισέρχεται σε ένα χώρο στον οποίο συχνότατα έχουν χάσει τη ζωή τους, είτε τραυματιστεί πολύ σοβαρά ένας ή περισσότεροι άνθρωποι Τα συντρίμμια μιλάνε και εσύ ως διερευνητής καλείσαι με καθαρό μυαλό να εντοπίσεις τις αιτίες για τις απώλειες έμψυχες και υλικές. Και από την άλλη υπάρχει και το ζήτημα των απειλών κατά της υγείας των μελών της ομάδας έρευνας για παράδειγμα από ραδιενεργά, τοξικά η ιατροφαρμακευτικά υλικά που μπορεί να μεταφέρει το υπό διερεύνηση αεροσκάφος. Θα πρέπει εδώ να τονίσουμε ότι μεγάλη σημασία δίνεται από την επιτροπή όταν το αεροπλάνο μεταφέρει ταχυδρομείο το οποίο έχει επίσης προτεραιότητα στην επιχείρηση διάσωσης. Για να ξεκινήσει η διερεύνηση εκτός από την ασφάλιση των συντριμμιών, την καταγραφή και φωτογράφιση του χώρου όπου σημειώθηκε το ατύχημα ένας από τους διερευνητές θα ασχοληθεί με την εύρεση αυτοπτών μαρτύρων».
Αναφερόμενος σε μία επιπλέον «στενωπό» την οποία αντικρίζουν- αντιμετωπίζουν στην εκτέλεση του έργου τους οι ουκ ολίγες φορές συνάδελφοι του στις διερευνητικές ομάδες ο Ηλίας Τάρλας δηλώνει κατηγορηματικά: «Κάθε αεροπορικό ατύχημα έχει να κάνει με μεγάλα ή… μεγαλύτερα οικονομικά συμφέροντα. Στην περίπτωση που έχουμε κάποιο επιβατικό αεροσκάφος θα κληθεί η εταιρεία να πληρώσει αποζημιώσεις ενώ κάτι ανάλογο μπορεί να γίνει και τόσο από πλευράς ιδιωτών όσο και του κράτους. Όπως καταλαβαίνετε ορισμένες φορές υπάρχει μια, ας την ορίσουμε επιεικώς , συγκρατημένη υποδοχή των διερευνητικών ομάδων από τους εμπλεκόμενους ή έχοντες συμφέροντα σχετιζόμενα με το αεροπορικό ατύχημα».



### Που διαβιβάζονται οι εκθέσεις/πορίσματα
Με την ολοκλήρωση των διεργασιών διερεύνησης, η Ομάδα συντάσσει Σχέδιο Πορίσματος το οποίο υποβάλλει στην Επιτροπή που το εξετάζει και εκδίδει το Τελικό Πόρισμα το οποίο διαβιβάζεται στον Υπουργό Υποδομών & Μεταφορών, σε όλους τους φορείς και τις υπηρεσίες οι οποίες εμπλέκονται άμεσα ή έμμεσα στο ατύχημα/συμβάν, στον ICAO (Διεθνής Οργανισμός Πολιτικής Αεροπορίας), την Ευρωπαϊκή Επιτροπή και εφόσον από το διερευνηθέν συμβάν ή ατύχημα έχει προέλθει θάνατος ή βαριά σωματική βλάβη, στην αρμόδια Εισαγγελική Αρχή ενώ όλα τα στοιχεία που χρησιμοποιήθηκαν για τη διερεύνηση παραμένουν στην κατοχή της ΕΔΑΑΠ που εκδίδει επί συμβουλευτικής βάσης Συστάσεις Ασφαλείας Πτήσεων προς τους Εθνικούς και Διεθνείς αρμόδιους φορείς, προκειμένου να διορθωθούν αδυναμίες και ελλείψεις που εντοπίζονται κατά τις διερευνήσεις ή προκύπτουν από την παγκόσμια αεροπορική εμπειρία. Επίσης ασχολείται με την τήρηση και επεξεργασία στατιστικών στοιχείων ατυχημάτων και συμβάντων σε εθνική και διεθνή βάση και την παρακολούθηση της εφαρμογής των συστάσεων ασφαλείας των πορισμάτων. Συνεργάζεται στενά με τις αντίστοιχες επιτροπές των άλλων κρατών μελών του ICAO καθώς και με κάθε Αρχή, Οργανισμό ή Ίδρυμα που ασχολείται με την ασφάλεια των πτήσεων και οι τελικές Εκθέσεις Διερεύνησης και Συστάσεις Ασφαλείας που εκδίδει δημοσιοποιούνται. 

















### To πλέον… χρονοβόρο πόρισμα 
Η έκδοση Συστάσεων Ασφαλείας Πτήσεων που απορρέουν από την διερεύνηση ενός ατυχήματος ή συμβάντος αποτελεί τον πρωταρχικό στόχο του έργου της Επιτροπής. Το περιεχόμενο μιας Σύστασης Ασφάλειας συνδέεται με τα αίτια του συγκεκριμένου ατυχήματος ή του συμβάντος και δεν συνιστά, σε καμιά περίπτωση, τεκμήριο υπαιτιότητας ή ευθύνης. Οι περισσότερες Συστάσεις εκδίδονται με την ολοκλήρωση του έργου της διερεύνησης και περιέχονται στο Πόρισμα. Κατ' εξαίρεση, Συστάσεις Ασφάλειας μπορούν να εκδοθούν και πριν ολοκληρωθεί η διερεύνηση, εφόσον, κατά την κρίση της Επιτροπής, υφίσταται άμεσος κίνδυνος μελλοντικών ατυχημάτων ή συμβάντων από ίδια ή παρόμοια αίτια. Οι Συστάσεις Ασφάλειας απευθύνονται στους φορείς που θεωρείται ότι πρέπει να λάβουν τα προσήκοντα για την περίπτωση διορθωτικά μέτρα.
Ένα από τα αξιοσημείωτα στοιχεία που προκύπτουν από την έρευνα της δράσης της ΕΔΑΑ είναι ο απαιτούμενος χρόνος για τη σύνταξη των εκθέσεων /πορισμάτων της. Κατά μέσο όρο αυτός είναι γύρω στα δύο χρόνια αν και υπάρχουν περιπτώσεις που κατατέθηκαν σε διάστημα λίγων μηνών ενώ στον αντίποδα σε τρεις περιπτώσεις αυτό έγινε έπειτα από διάστημα τεσσάρων και πλέον ετών αν και αξίζει να αναφερθεί ουδεμία αφορούσε ατύχημα με σοβαρό ή θανάσιμο τραυματισμό. Το ρεκόρ … καθυστέρησης έκδοσης έχει  με διάστημα εννιά (!) ετών το τελικό πόρισμα αναφορικά με το ατύχημα δίχως τραυματισμό που σημειώθηκε στις 8/11/2013 στο αεροδρόμιο Μεγάρων όταν ελαφρύ αεροσκάφος κατά τη διάρκεια προσγείωσης αναπήδησε μετά από βαριά προσγείωση, εξετράπη και συνετρίβη σε κατασκευή ISOBOX. Μάλιστα όπως μας είπε χαρακτηριστικά ο  σχεδόν μισό αιώνα πιλότος (πρώτη πτήση το 1978 και στο επάγγελμα 39 συναπτά έτη) «είναι απίστευτο που στο συγκεκριμένο ατύχημα αμφότεροι οι χειριστές βρήκαν από  ότι απέμεινε από το αεροπλάνο ατσαλάκωτοι». Η έκδοση έγινε στις 10 του περασμένου Νοεμβρίου. Στον αντίποδα το τελικό πόρισμα για  το σοβαρό συμβάν δίχως τραυματισμό που σημειώθηκε στις 27 Νοεμβρίου του 2004 στον Κρατικό Αερολιμένα Ηρακλείου σε σύγκρουση έλικα μεγάλου ελικοπτέρου με κολώνα ηλεκτρικού φωτισμού κατά την τροχοδρόμηση εκδόθηκε στις 9 Μαρτίου 2005 δηλαδή σχεδόν 4,5 μήνες αργότερα.
Κύριος αποδέκτης των συστάσεων της Επιτροπής είναι η Υπηρεσία Πολιτικής Ασφαλείας ενώ το ίδιο συμβαίνει σε ουκ ολίγες περιπτώσεις και με αερομεταφορείς, κατασκευαστές και οργανισμούς (αεραθλητική ομοσπονδία, ΔΕΗ κ.α). Η σύσταση είναι είτε ανοικτή, όταν προβλέπεται επιπλέον διάλογος με τον αποδέκτη για ενέργειες σχετικά με αυτή, είτε κλειστή σε αντίθετη περίπτωση.

Κατόρθωσα και κράτησα το αεροπλάνο στον διάδρομο και διαπιστώθηκαν έπειτα από διερεύνηση τα εξής: 1) ο τροχός δεν έφυγε μόνος του 2) στη συντήρηση των φρένων οι τεχνικοί είχαν βγάλει τον τροχό 3) στην επανατοποθέτηση δεν ασφάλισαν τον τροχό με το ειδικό εργαλείο και 4) κατασκευαστική έλλειψη. Στη συγκεκριμένη περίπτωση έχοντας μακρά πείρα πτήσεων κατά τη διάρκεια της εξέλιξης της προσγείωσης ενήργησα βάση της εκπαίδευσης μου απόλυτα μηχανικά αποκλείοντας τη συναισθηματική εμπλοκή. Το τι συνέβη το αντιλήφθηκα μετά την ακινητοποίηση του αεροπλάνου. Και ευκαιρίας δοθείσης να τονίσω ότι η εκπαίδευση του χειριστή πάνω από το 50% αφορά στις διαδικασίες ανάγκης, Δηλαδή εκπαιδευόμαστε σε διάφορα σενάρια που μπορούν να συμβούν στον αέρα. Το πέταγμα του αεροπλάνου είναι πάρα πολύ εύκολο. Μπορώ να πω ότι είναι ευκολότερο και από ποδήλατο, Απλά πρέπει να εφαρμόζεις πιστά ότι περιγράφεται στο check – list στην ανάλογη διαδικασία ανάγkης. Άρα οφείλουμε να ενεργούμε ψύχραιμα και μηχανικά για την εκτέλεση των διαδικασιών και αυτό έπραξα και στη συγκεκριμένη σοβαρή «κρίση» την οποία αντιμετώπισα πριν από περίπου 3,5 χρόνια. ### Η ανεξαρτησία και η διεθνής δράση της ΕΔΑΑΠ
Η ΕΔΑΑΠ, που σύμφωνα με την Οδηγία 94/56 της Ευρωπαϊκής Κοινότητας θεσπίστηκε ως «οργανισμός ή όργανο διερεύνησης αεροπορικών ατυχημάτων λειτουργικά ανεξάρτητο από τις εθνικές αεροπορικές αρχές τις υπεύθυνες για την πλοϊμότητα, την πιστοποίηση, τη διεξαγωγή πτήσεων, τη συντήρηση, τη χορήγηση αδειών, τον έλεγχο της εναέριας κυκλοφορίας ή την εκμετάλλευση αερολιμένων και γενικά από κάθε τρίτο του οποίου τα συμφέροντα θα μπορούσαν να έρθουν σε σύγκρουση με την αποστολή που ανατέθηκε στον εν λόγω οργανισμό ή όργανο», επίσης εκπροσωπεί την χώρα μας σε διερευνήσεις οι οποίες διεξάγονται από ξένο κράτος στην περίπτωση ατυχήματος ή συμβάντος αεροσκάφους με ελληνικό νηολόγιο εκτός του FIR Αθηνών. για τη «θέσπιση βασικών αρχών που διέπουν τις έρευνες ατυχημάτων και συμβάντων της Πολιτικής Αεροπορίας». Σημερινός της πρόεδρος είναι ο διερευνητής ατυχημάτων και συμβάντων, τέως κυβερνήτης Πολιτικής Αεροπορίας Ιωάννης Κωνσταντίνος Κονδύλης και μέλη πέραν του αντιπροέδρου Ακριβού Τσολάκη,  οι Γρηγόριος Φλέσσας (απόστρατος Π.Α & κυβερνήτης Πολιτικής Αεροπορίας), Χρήστος Βάλαρης (ταξίαρχος ε.α Π.Α) και Χαράλαμπος τζώνος Κομίλης (κυβερνήτης Π.Α.) Διαδέχθηκε την Ανακριτική Επιτροπή Αεροπορικών Ατυχημάτων (ΑΕΑΑ), το Συμβούλιο Ερεύνης Αεροπορικών Ατυχημάτων (ΣΕΑΑ) και το Ανακριτικό Συμβούλιο Αεροπορικών Ατυχημάτων (ΑΣΑΑ) που είχαν μετά τον Β’ Παγκόσμιο Πόλεμο και ως τα τέλη του προηγούμενου αιώνα, σύμφωνα με την ελληνική νομοθεσία, τη σχετική αρμοδιότητα και αποτελούντο από υπαλλήλους της Υπηρεσίας Πολιτικής Αεροπορίας (ΥΠΑ) ή αξιωματικούς της Ελληνικής Βασιλικής Αεροπορίας (ΕΒΑ) και το έργο της είναι ανοικτό σε κάθε ενδιαφερόμενο (ιδιώτη, οργανισμό, δημόσιο) στην επίσημη διαδικτυακή της πύλη (https://aaiasb.gr/).
