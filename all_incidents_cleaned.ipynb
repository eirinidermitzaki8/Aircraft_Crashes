```python
# imports
import pandas as pd
# setups
pd.set_option('display.max_rows', None)
```

```python
# reading our df with the data we scraped previously
df = pd.read_csv('all_incidents.csv')
```

```python
# checking our df's head
df.head()
```

```python
# turning conclusion_date2 into a datetime object, because it is a string currently
df['conclusion_date2'] = pd.to_datetime(df.conclusion_date2)
```

```python
# using some simple regex to clean incident_info in order to keep dates only
# then, turn it into a datetime object
df['incident_info'] = df.incident_info.str.extract(r'(\d+\/\d+\/\d+)')
df['incident_info'] = pd.to_datetime(df.incident_info)
```

```python
# renaming columns, so as to be more accurate
df = df.rename(columns = 
               {'conclusion_date1':'conclusion_number',
                'conclusion_date2': 'conclusion_date', 
                'incident_info' :'incident_date'})
```

```python
# cleaning incident variables and fatalities 
df['incident_type'] = df.incident_type.str.strip()
df['incident_description'] = df.incident_description.str.replace('<span span="" uk-icon="info" uk-tooltip="','')
df['incident_description'] = df.incident_description.str.replace('"><br/></span>','').str.strip()
df['fatalities'] = df.fatalities.str.strip()
df.head()
```

```python
# creating a variable called GARBAGE and we will put inside '()' different stringswe want to remove.
GARBAGE = '(<span <="" span="" uk-icon="info" uk-tooltip="|\xa0|">\n</span>|&lt;br&gt;)'

# now we are gonna replace the variable GARBAGE with nothing. 
df['fatalities_description'] = df.fatalities_description.str.replace(GARBAGE, '', regex=True)
df.head()
```

```python
# now we are gonna use some regex in order to extract the number of the crew members, the number of deaths, the number of injuries etc
#we will store those data in corresponding new columns
df['crew'] = df.fatalities_description.str.extract(r'Πλήρωμα=(\d+)')
df['deaths_crew_members'] = df.fatalities_description.str.extract(r'Θάνατοι\s+\Πληρώματος=(\d+)')
df['seriously_injured_crew_members'] = df.fatalities_description.str.extract(r'Σοβ\.\s+Τραυμ.\s+πληρώματος=(\d+)')
df['passengers'] = df.fatalities_description.str.extract(r'Επιβάτες=(\d+)')
df['deaths_passengers'] = df.fatalities_description.str.extract(r'Θάνατοι\s+\Επιβατών=(\d+)')
df['seriously_injured_passengers'] = df.fatalities_description.str.extract(r'Σοβ\.\s+Τραυμ.\s+επιβατών=(\d+)')
df['other_deaths'] = df.fatalities_description.str.extract(r'Εκτός\s+\αεροσκάφους\s+άλλοι\s+θάνατοι=(\d+)')
df['other_injuries'] = df.fatalities_description.str.extract(r'άλλοι\s+Σοβ\.\s+Τραυμ\.=(\d+)')
```

```python
df.head()
```

```python

# using some  regex to estract registry info found in the area_info column and store it in a new col named registry2
df['registry2'] = df.area_info.str.extract(r'Νηολόγιο\(α\):(.*)Α|Υ|Ε|Μ')
# we will fill empty entries in our major registry column with what extracted from the area_info and is temporarily stored in the registry2 column
df['registry'] = df.registry.fillna(df.registry2).str.strip()
# some entries have '-' as a registry value. We can replace that with NaN, because either way they are missing values
df['registry'] = df.registry.str.replace('^(-|-)$','',regex=True).replace('',float('NaN'))
# cleaning registry further
df['registry'] = df.registry.str.replace('Νηολόγιο\(α\):','')
df['registry'] = df.registry.str.replace('Μηχανοκίνητο','')
df['registry'] = df.registry.str.strip()
# droping the registry2 column we created for temporary use 
df = df.drop(['registry2'],axis=1)
```

```python
df.head()
```

```python
# cleaning aircraft_type and filling missing values with info that can be extracted from the area_info column
TYPE = r'(Αεροπλάνο|Αλεξίπτωτο|Ελικόπτερο|Ανεμόπτερο|Αιωρόπτερο)'
df['aircraft_type'] = df.aircraft_type.replace('Νηολόγιο\(α\):.*',float('NaN'),regex=True)
df['aircraft_type2'] = df.area_info.str.extract(TYPE)
df['aircraft_type'] = df.aircraft_type.fillna(df.aircraft_type2).str.strip()
```

```python
# applying the same methods to fill and clean area data
df['area2'] = df.area_info.str.extract(r'(.*)Νηολόγιο')
df['area'] = df.area.fillna(df.area2).str.title().str.strip()
df = df.drop(['area2'],axis=1)
df.head()
```

```python
# saving our cleaned dataset
df.to_csv('all_incidents_cleaned.csv',index=False)
```
